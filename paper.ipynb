```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from io import StringIO
from matplotlib.colors import LinearSegmentedColormap

plt.rcParams['font.family'] = 'Times New Roman'

df = pd.read_csv(StringIO(data_tsv), sep='\t')
df.columns = ['early_feedback','resubmission','process','ai_comfort','trust_appeal','role_clarity','assess_conf']

labels_short = {1:'SD', 2:'D', 3:'N', 4:'A', 5:'SA'}
full_labels = {1:'Strongly\ndisagree', 2:'Disagree',
               3:'Neutral', 4:'Agree', 5:'Strongly\nagree'}

pairs = [
    ('ai_comfort',   'trust_appeal',  'Comfort with AI-assisted feedback',  'Trust in challenge/appeal path'),
    ('ai_comfort',   'role_clarity',  'Comfort with AI-assisted feedback',  'Clarity of AI vs. TA/instructor role'),
    ('role_clarity', 'assess_conf',   'Clarity of AI vs. TA/instructor role','Confidence in TA regrading fairness'),
]

FS = 22
fig, axes = plt.subplots(1, 3, figsize=(24, 8))
fig.subplots_adjust(wspace=0.55)
cmap = plt.cm.Purples

cmap = LinearSegmentedColormap.from_list(
    "custom_blue",
    ["#F2F6FA", "#92BFE5", "#4884BC", "#1F5A99"]
)

for ax, (row_col, col_col, row_label, col_label) in zip(axes, pairs):
    ct = pd.crosstab(df[row_col], df[col_col])
    all_vals = [1,2,3,4,5]
    ct = ct.reindex(index=all_vals, columns=all_vals, fill_value=0)
    ct_prop = ct.div(ct.sum(axis=1), axis=0).fillna(0)

    rows = ct_prop.index.tolist()
    cols = ct_prop.columns.tolist()
    mat  = ct_prop.values

    im = ax.imshow(mat[::-1], cmap=cmap, vmin=0, vmax=1, aspect='equal')

    # annotate — skip 0.00
    for i in range(len(rows)):
        for j in range(len(cols)):
            val = mat[len(rows) - 1 - i, j]
            if val == 0:
                continue
            color = 'white' if val > 0.55 else 'black'
            ax.text(j, i, f'{val*100:.0f}%', ha='center', va='center',
                    fontsize=FS, color=color,
                    fontweight='bold' if val > 0.4 else 'normal',
                    zorder=4)

    ax.set_xticks(range(len(cols)))
    ax.set_xticklabels([labels_short[c] for c in cols], fontsize=FS)
    ax.set_yticks(range(len(rows)))
    ax.set_yticklabels([full_labels[r] for r in rows[::-1]], fontsize=FS)
    ax.set_xlabel(col_label, fontsize=FS, labelpad=10, weight='bold')
    ax.set_ylabel(row_label, fontsize=FS, labelpad=10, weight='bold')

    # light grid lines
    for x in np.arange(-0.5, len(cols), 1):
        ax.axvline(x, color='white', linewidth=1.2, zorder=5)
    for y in np.arange(-0.5, len(rows), 1):
        ax.axhline(y, color='white', linewidth=1.2, zorder=5)
    ax.tick_params(length=0)
    for spine in ax.spines.values():
        spine.set_visible(False)

# shared colorbar — horizontal at bottom
cbar_ax = fig.add_axes([0.25, -0.1, 0.5, 0.025])
sm = plt.cm.ScalarMappable(cmap=cmap, norm=mcolors.Normalize(vmin=0, vmax=1))
sm.set_array([])
cb = fig.colorbar(sm, cax=cbar_ax, orientation='horizontal')
cb.set_label('Row proportion (%)', fontsize=FS, labelpad=10)
cb.ax.xaxis.set_label_position('top')
cb.ax.tick_params(labelsize=FS)
cb.set_ticks([0, 0.25, 0.5, 0.75, 1.0])
cb.set_ticklabels(['0%', '25%', '50%', '75%', '100%'])
cb.outline.set_edgecolor('white')
cb.outline.set_linewidth(1.0)

fig.suptitle('Cross-tabulations between trust and transparency items ($N=95$)',
             fontsize=FS*1.2)
fig.tight_layout(rect=[0, 0.1, 1, 0.96])
fig.subplots_adjust(wspace=0.55)

# plt.savefig('/mnt/user-data/outputs/crosstab_heatmaps.pdf', bbox_inches='tight', dpi=300)
# plt.savefig('/mnt/user-data/outputs/crosstab_heatmaps.png', bbox_inches='tight', dpi=300)
# print("Done")
plt.show()
```
